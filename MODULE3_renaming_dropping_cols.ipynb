```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['columns']= (tips['tip']/tips['total_bill']).round(2)
tips.head()
```

```python
tips.rename(columns = {'column': 'tip_pct'},inplace = True)
tips.columns
```
