```python
import pandas as pd
import seaborn as sns
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
mpg['mpg'].plot()
```

```python
mpg['mpg'].plot(kind='hist')
```

```python
mpg['mpg'].plot(kind='box')
```

```python
mpg['mpg'].plot(kind='box',vert=False)
```

```python
mpg['origin'].value_counts()
```

```python
mpg['origin'].value_counts().plot(kind='barh')
```

```python
mpg['origin'].value_counts().plot(kind='pie')
```

```python
mpg.plot(x='weight',y='mpg',kind='scatter')
```
